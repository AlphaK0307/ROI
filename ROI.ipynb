```python
class ROI:
    def __init__(self):
        self.propertycost = None
        self.rent = None
        self.totalincomemonthly = None
        self.totalexpenses = None
        self.totalCashFlowmonth = None
        
    def income(self):
        self.propertycost = int(input("Listing price for the property: "))
        self.rent = int(input("Enter the monthly $$ you receive in rent from your tenants: "))
        otherincome = int(input("$$ Other income you'll recieve from tenants ex,Laundry,storage,misc: "))
        self.totalincomemonthly = self.rent + otherincome
        print(f"The total monthly incomeis: {self.totalincomemonthly}")

    def expenses(self):
        mortgage=int(input("What is your monthly mortgage?: "))
        tax = int(input("Property taxes you will have to pay: "))
        insurance = int(input("Insurance amount: "))
        self.totalexpenses = tax + insurance + mortgage
        print(f'Monthly expenses: {self.totalexpenses}')
    
    def CashFlow(self):
        self.totalMcashflow = self.totalincomemonthly - self.totalexpenses
        print(f"Monthly cash flow {self.totalMcashflow} after deducting {self.totalexpenses}")
    
    def investmentreturn(self):
        downpayment = int(input(" DownPayment on property: "))
        closingcost = int(input("Closing cost: "))
        fixingcost = int(input("Fixing costs if any?: "))
        Investment = downpayment + closingcost + fixingcost
        print(f"Total investment: {Investment}")
        annualcashflow = self.totalMcashflow * 12
        returnofinvestment = annualcashflow / Investment
        print(f"Your return on investment is {returnofinvestment}")

        
newROI = ROI()


def run():
    while True:
        calc = input("Calculate your: initial investment(1), exspenses(2), cash flow(3) or your return on investment(4) or 'end' to end.")
        if calc == "1":
            print(newROI.income())
        elif calc == "2":
            print(newROI.expenses())
        elif calc == "3":
            print(newROI.cashFlow())
        elif calc == "4":
            print(newROI.investmentreturn())

        elif calc == "end":
            break
run()
        

   
```
